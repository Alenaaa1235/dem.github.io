```python
import math

# Ввод двух чисел
a = float(input("Введите первое число: "))
b = float(input("Введите второе число: "))

# Проверка: числа должны быть неотрицательными
if a < 0 or b < 0:
    print("Ошибка: среднее геометрическое определено только для неотрицательных чисел!")
else:
    # Вычисление среднего геометрического
    geometric_mean = math.sqrt(a * b)
    print(f"Среднее геометрическое: {geometric_mean}")
```

```python
import math

# Ввод длин катетов
a = float(input("Введите длину первого катета: "))
b = float(input("Введите длину второго катета: "))

# Проверка на отрицательные значения
if a <= 0 or b <= 0:
    print("Ошибка: длины катетов должны быть положительными числами!")
else:
    # Вычисление гипотенузы (теорема Пифагора)
    hypotenuse = math.sqrt(a**2 + b**2)
    
    # Вычисление площади
    area = (a * b) / 2
    
    # Вывод результатов
    print(f"Гипотенуза: {hypotenuse}")
    print(f"Площадь: {area}")
```

```python
# Инициализация счетчика
even_count = 0

# Ввод 10 чисел
for i in range(1, 11):
    number = int(input(f"Введите число {i}: "))
    if number % 2 == 0:
        even_count += 1

# Вывод результата
print(f"Количество четных чисел: {even_count}")
```

```python
# Инициализация счетчиков
positive = 0
negative = 0
zeros = 0

# Ввод 10 чисел
for i in range(1, 11):
    number = int(input(f"Введите число {i}: "))
    if number > 0:
        positive += 1
    elif number < 0:
        negative += 1
    else:
        zeros += 1

# Вывод результатов
print(f"Положительных чисел: {positive}")
print(f"Отрицательных чисел: {negative}")
print(f"Нулей: {zeros}")
```

```python
# Ввод первого числа для инициализации
min_number = int(input("Введите число 1: "))
max_number = min_number

# Ввод остальных 9 чисел
for i in range(2, 11):
    number = int(input(f"Введите число {i}: "))
    if number < min_number:
        min_number = number
    if number > max_number:
        max_number = number

# Вычисление разницы
difference = max_number - min_number

# Вывод результатов
print(f"Наибольшее число: {max_number}")
print(f"Наименьшее число: {min_number}")
print(f"Разница: {difference}")
```

```python
# Ввод массива из 10 целых чисел
A = []

for i in range(10):
    num = int(input(f"Введите элемент A[{i}]: "))
    A.append(num)

# Находим максимальный элемент
max_element = max(A)

# Подсчитываем количество элементов, равных максимальному
count_max = A.count(max_element)

# Вывод результатов
print(f"Массив: {A}")
print(f"Максимальный элемент: {max_element}")
print(f"Количество элементов, равных максимальному: {count_max}")
```

```python
# Ввод массива из 10 целых чисел
A = []

for i in range(10):
    num = int(input(f"Введите элемент A[{i}]: "))
    A.append(num)

# Вычисление сумм
sum_even = 0  # сумма элементов с четными индексами (0, 2, 4, 6, 8)
sum_odd = 0   # сумма элементов с нечетными индексами (1, 3, 5, 7, 9)

for i in range(10):
    if i % 2 == 0:
        sum_even += A[i]
    else:
        sum_odd += A[i]

# Вычисление разницы
difference = sum_even - sum_odd

# Вывод результатов
print(f"Массив: {A}")
print(f"Сумма элементов с четными индексами: {sum_even}")
print(f"Сумма элементов с нечетными индексами: {sum_odd}")
print(f"Разница (четные - нечетные): {difference}")
```

```python
print("Игра 'Угадай число'")
print("Загадайте число из заданного диапазона, а компьютер отгадает его!")
print()

# Ввод диапазона
N = int(input("Введите нижнюю границу диапазона N: "))
M = int(input("Введите верхнюю границу диапазона M: "))

# Проверка корректности диапазона
if N > M:
    print("Ошибка: нижняя граница больше верхней!")
    exit()

print(f"\nЗагадайте число от {N} до {M}.")
print("Компьютер будет задавать вопросы типа 'да' или 'нет'.")
print("Отвечайте: 'да' (или 'д'), 'нет' (или 'н')")
print()

# Начальные границы поиска
left = N
right = M
attempts = 0

while left < right:
    # Берем среднее значение
    mid = (left + right) // 2
    
    # Задаем вопрос
    attempts += 1
    answer = input(f"Ваше число больше {mid}? (да/нет): ").lower()
    
    if answer in ['да', 'д', 'yes', 'y']:
        left = mid + 1
    elif answer in ['нет', 'н', 'no', 'n']:
        right = mid
    else:
        print("Пожалуйста, отвечайте 'да' или 'нет'!")
        attempts -= 1  # не считаем некорректный ответ

# Компьютер угадал число
print(f"\nКомпьютер угадал ваше число: {left}!")
print(f"Количество вопросов: {attempts}")
```
